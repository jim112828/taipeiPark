```python
import pandas as pd

import json
```

```python
data = json.load(open("apiIn.json"))
```

```python
df = pd.DataFrame(data)
df
```

```python
df=df[["Image","ParkName","Latitude","Longitude"]]
```

```python
df["Latitude"][0]
```

```python
import folium
taipeiMap = folium.Map(location=[25.05, 121.33],zoom_start=10)
html = """<h1>"""+df["ParkName"][0]+"""</h1><br><img src="""+df["Image"][0]+""">"""


for i in range(len(df)):
    html = """<h1>"""+df["ParkName"][i]+"""</h1><br><img src="""+df["Image"][i]+""" width="300" height="300">"""
    folium.Marker([float(df["Latitude"][i]), float(df["Longitude"][i])], popup=html).add_to(taipeiMap)

taipeiMap.save("taipeiPark.html")

```
